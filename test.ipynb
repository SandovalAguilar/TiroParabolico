```python
import numpy as np
import matplotlib.pyplot as plt
import math

# theta=pi/3
V = 38.5 
t = np.arange(0, 100, 0.1)
angulo = math.pi / 4
angulo_lista = []

while angulo < math.pi / 3:
    x = V * np.cos(angulo) * t
    y = V * np.sin(angulo) * t + (0.5 * -9.80665 * t ** 2)
    plt.ylim(0, 80)
    plt.xlim(0, 170)
    plt.plot(x, y)
    zero_crossing_idx = np.argwhere(y > 0)[-1]
    angulo_lista.append(rf'{np.rad2deg(angulo):.0f}$\degree$')
    angulo += math.pi / 12

plt.xlabel("Alcance")
plt.ylabel("Altura")
plt.title('Trayectoria de proyectiles con múltiples ángulos')
plt.legend(angulo_lista)
plt.show()
```
